```python
import numpy as np
import matplotlib.pyplot as plt
```

some overal parameters

```python
r = 5
interval = 0.1
x = np.arange(-r,r+1, interval)
```

## ReLU

```python
relu = np.where(x > 0, x, 0)
drelu = np.where(relu > 0, 1, 0)
```

```python
rp = 5.5
plt.plot(x, relu, color='blue')
plt.plot(x, drelu, color='red', alpha = 0.5)
plt.title('ReLU')
plt.ylim(-rp,rp)
plt.xlim(-rp,rp)
plt.savefig('relu.pdf')
```

## Tanh

```python
tanh = np.tanh(x)
dtanh = 1-tanh**2
```

```python
ylim = 1.5
plt.plot(x, tanh, color='blue')
plt.plot(x, dtanh, color='red')
plt.title('tanh')
plt.ylim(-ylim,ylim)
plt.xlim(-rp,rp)
plt.savefig('tanh.pdf')
```

## sigmoid

```python
sigmoid = 1/(1+np.exp(-x))
dsigmoid = sigmoid*(1-sigmoid)
```

```python
ylim = 1.5
plt.plot(x, sigmoid, color='blue')
plt.plot(x, dsigmoid, color='red')
plt.title('sigmoid')
plt.ylim(-ylim,ylim)
plt.xlim(-rp,rp)
plt.savefig('sigmoid.pdf')
```
